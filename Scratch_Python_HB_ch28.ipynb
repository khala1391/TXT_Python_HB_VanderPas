### Visualizing 3D function 

```python
def f(x,y):
    return np.sin(X)**10 + np.cos(10 + y*x)*np.cos(x)
```

```python
x = np.linspace(0,5,50)
y = np.linspace(0,5,40)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
```

```python
X.shape, Y.shape, Z.shape
```

```python
plt.contour(X,Y,Z, colors='black');
```

```python
plt.contour(X,Y,Z,20, cmap='RdGy')
plt.colorbar();
```

```python
plt.contourf(X,Y,Z,20, cmap='RdGy')
plt.colorbar();
```

```python
plt.imshow(Z,extent=[0,5,0,5], origin='lower', cmap='RdGy',
          interpolation='gaussian',aspect='equal')
plt.colorbar();
```

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
cmap='RdGy', alpha=0.5)

plt.colorbar();
```

### Histogram, Binnings, and Density 

```python
rng = np.random.default_rng(1701)
data = rng.normal(size=1000)
```

```python
plt.hist(data);
```

```python
plt.hist(data,bins=30, density=True, alpha=0.5,
        histtype='barstacked', color='steelblue',
        edgecolor='none');
```

```python
x1 = rng.normal(0, 0.8, 1000)
x2 = rng.normal(-2, 1, 1000)
x3 = rng.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, density=True, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

```python
# Equivalent

x1 = rng.normal(0, 0.8, 1000)
x2 = rng.normal(-2, 1, 1000)
x3 = rng.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, density=True, bins=40)

plt.hist(x1, histtype='stepfilled', alpha=0.3, density=True, bins=40)
plt.hist(x2, histtype='stepfilled', alpha=0.3, density=True, bins=40)
plt.hist(x3, histtype='stepfilled', alpha=0.3, density=True, bins=40);
```
